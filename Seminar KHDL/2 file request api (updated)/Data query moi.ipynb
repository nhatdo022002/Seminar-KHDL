# ghi dữ liệu vào file Json và lưu về máy

```python
import json
import requests
import pandas as pd

url = "https://api.themoviedb.org/3/movie/upcoming?language=en-US&page=1"

headers = {
    "accept": "application/json",
    "Authorization": "Bearer eyJhbGciOiJIUzI1NiJ9.eyJhdWQiOiI3ODFkOGIwZjMzMTYzMTgxZDk2ZDYxZmEyMWVhMTZhNCIsInN1YiI6IjY1MWRhNmVkYzUwYWQyMDBhZDgyZDRlYSIsInNjb3BlcyI6WyJhcGlfcmVhZCJdLCJ2ZXJzaW9uIjoxfQ.jfoB836ahuG6GYsOsAOTORYACLJRO6V7mm0_IUxMLv4"
}

response = requests.get(url, headers=headers)
if response.status_code == 200:
    # Chuyển đổi nội dung JSON thành đối tượng Python
    json_data = json.loads(response.text)

    # Lưu JSON vào một tệp văn bản
    with open("output_upcoming_id.json", "w") as file:
        json.dump(json_data, file, indent=4)
else:
    print("Lỗi khi tải dữ liệu từ API")
#print(response.text)
```

## chuyển genres_id sang tên (dùng để truy vấn sql)

```python
id_name_map = genres_data.set_index('id')['name'].to_dict()
# Hàm chuyển đổi ID thành Tên thể loại trong danh sách
def convert_id_list_to_name_list(genres_ids):
    return [id_name_map[id] for id in genres_ids]

# Áp dụng hàm chuyển đổi cho mỗi dòng trong cột genres_id
df['genre_ids'] = df['genre_ids'].apply(convert_id_list_to_name_list)
# Hiển thị kết quả
df
```
